```python
import pandas as pd
df = pd.read_csv('/content/spam.csv', encoding='latin-1')
df= df.iloc[:, 0:2]
df.head()
```

```python
df['v1'].replace({'ham':1,'spam':0},inplace =True)
df.head()
```

```python
messages =df['v2']
```

```python
from nltk.stem import PorterStemmer
import re
pc = PorterStemmer()
```

```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
```

```python
processed_messages = []
messages_series = df['v2'] # Get the 'v2' column as a Series
messages_list = messages_series.tolist() # Convert Series to list

for message in messages_list:
  review = re.sub('[^a-zA-Z]',' ',message)
  review =review.lower()
  review = review.split()
  review =[pc.stem(word) for word in review if word not in stopwords.words('english')]
  processed_messages.append(' '.join(review))

df['v2'] = processed_messages
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=100,ngram_range=(1,2))
X = cv.fit_transform(df['v2']).toarray() # X now contains the Bag-of-Words features
```

```python
Y = df['v1']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
nb = MultinomialNB()
nb.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

y_pred = nb.predict(X_test)
print(accuracy_score(Y_test,y_pred)*100)
print(confusion_matrix(Y_test,y_pred))
print(classification_report(Y_test,y_pred))
```
